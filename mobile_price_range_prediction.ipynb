```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/mobile_price_range_data (1).csv")
df.shape
```

```python
ax=sns.boxplot(data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
```

```python
def outlier(df, col):
    iqr=df[col].quantile(0.75) - df[col].quantile(0.25)
    max=df[col].quantile(0.75) + 1.5*iqr
    min=df[col].quantile(0.25) - 1.5*iqr
    df_out=df.loc[(df[col]>min) & (df[col]<max)]
    return(df_out)

num_col=['px_height','fc','ram','sc_w']
for i in num_col:
    df=outlier(df,i)
```

```python
df.shape
```

```python
from sklearn import metrics , svm
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics._plot.confusion_matrix import confusion_matrix
```

```python
x= df.drop('price_range',axis=1)
y= df['price_range']
```

```python
# train test split
x_train, x_test,y_train, y_test = train_test_split(x,y, random_state=10,test_size=0.25,shuffle=True)
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train, y_train)
y_pred1= knn.predict(x_test)
```

##KNN

```python
# acuuracy , confusion metrix and classification_report of KNN
accuracy_KNN=np.round(metrics.r2_score(y_test,y_pred1),2)
confusion_KNN= pd.DataFrame(confusion_matrix(y_test,y_pred1))
print(classification_report(y_test,y_pred1))
```

##SVC

```python
svc = svm.SVC(kernel='linear', C=1).fit(x_train, y_train)
y_pred2=svc.predict(x_test)
```

```python
# acuuracy , confusion metrix and classification_report of SVC
accuracy_SVC=np.round(metrics.r2_score(y_test,y_pred2),2)
confusion_SVC= pd.DataFrame(confusion_matrix(y_test,y_pred2))
print(classification_report(y_test,y_pred2))
```

##LOSITIC REGRESSION

```python
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)
```

```python
y_pred3 = classifier.predict(x_test)
```

```python
# acuuracy , confusion metrix and classification_report of LogisticRegression
accuracy_LOG=np.round(metrics.r2_score(y_test,y_pred3),2)
confusion_LOG= pd.DataFrame(confusion_matrix(y_test,y_pred3))
print(classification_report(y_test,y_pred3))
```

##DECISION TREE

```python
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
y_pred4 = model.predict(x_test)
```

```python
# acuuracy , confusion metrix and classification_report of DecisionTreeClassifier
accuracy_TREE=np.round(metrics.r2_score(y_test,y_pred4),2)
confusion_TREE= pd.DataFrame(confusion_matrix(y_test,y_pred4))
print(classification_report(y_test,y_pred4))
```

##RANDOM FOREST CLASSIFIER

```python
rndfor = RandomForestClassifier()
rndfor.fit(x_train, y_train)
y_pred5=rndfor.predict(x_test)
```

```python
# acuuracy , confusion metrix and classification_report of RandomForestClassifier
accuracy_FOREST=np.round(metrics.r2_score(y_test,y_pred5),2)
confusion_FOREST= pd.DataFrame(confusion_matrix(y_test,y_pred5))
print(classification_report(y_test,y_pred5))
```

```python
# ACCURACY SCORE COMPARISON
list=np.array([accuracy_KNN,accuracy_SVC,accuracy_LOG,accuracy_TREE,accuracy_FOREST])
lis2=np.array(['KNN','SVC','LOG','TREE','FOREST'])
dict = {B : A for A, B in zip(list,lis2)}
print(dict)
```
